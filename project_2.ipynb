```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
#Read in csv files
MHS_2015 = Path('MHS_2015totals.csv')
MHS_2016 = Path('MHS_2016totals.csv')
MHS_2017 = Path('MHS_2017totals.csv')

df_2015 = pd.read_csv(MHS_2015)
df_2016 = pd.read_csv(MHS_2016)
df_2017 = pd.read_csv(MHS_2017)
```

```python
df_2015.columns
```

```python
#Drop the columns/rows not needed
clean_2015 = df_2015[['Player', 'Ace', 'Err', 'Srv', 'RcvCount', 'Dig', 'Err.1', 'Kil', 'Err.2', 'Att', 'RcvErr',
        'RcvQualTot', 'GamesPlayed']]

clean_2015 = clean_2015.dropna(how='any')
clean_2015
```

```python
clean_2016 = df_2016[['Player', 'Ace', 'Err', 'Srv', 'RcvCount', 'Dig', 'Err.1', 'Kil', 'Err.2', 'Att', 'RcvErr',
        'RcvQualTot', 'GamesPlayed']]

clean_2016 = clean_2016.dropna(how='any')
clean_2016
```

```python
clean_2017 = df_2017[['Player', 'Ace', 'Err', 'Srv', 'RcvCount', 'Dig', 'Err.1', 'Kil', 'Err.2', 'Att', 'RcvErr',
        'RcvQualTot', 'GamesPlayed']]

clean_2017 = clean_2017.dropna(how='any')
clean_2017
```

```python
#Merge 2015 and 2016
df_1516 = pd.merge(clean_2015, clean_2016, on = 'Player', how = 'outer')
df_1516
```

```python
#Merge in 2017
final_df = pd.merge(df_1516, clean_2017, on = 'Player', how = 'outer')
final_df
```

```python
final_df.columns
```

```python
final_df.dtypes
```

```python
#Convert object dtypes to float
converted_data = final_df[['Ace_x', 'Err_x', 'Srv_x', 'RcvCount_x', 'Dig_x', 'Err.1_x',
       'Kil_x', 'Err.2_x', 'Att_x', 'RcvErr_x', 'RcvQualTot_x',
       'GamesPlayed_x', 'Ace_y', 'Err_y', 'Srv_y', 'RcvCount_y', 'Dig_y',
       'Err.1_y', 'Kil_y', 'Err.2_y', 'Att_y', 'RcvErr_y', 'RcvQualTot_y',
       'GamesPlayed_y', 'Ace', 'Err', 'Srv', 'RcvCount', 'Dig', 'Err.1', 'Kil',
       'Err.2', 'Att', 'RcvErr', 'RcvQualTot', 'GamesPlayed']].astype(float)
```

```python
Players = final_df['Player']


Aces = final_df['Ace_x']+final_df['Ace_y']+final_df['Ace']
Srv_err = final_df['Err_x']+final_df['Err_y']+final_df['Err']
Total_srv = final_df['Srv_x']+final_df['Srv_y']+final_df['Srv']
RcvCount = final_df['RcvCount_x']+final_df['RcvCount_y']+final_df['RcvCount']
Dig = final_df['Dig_x']+final_df['Dig_y']+final_df['Dig']
Defense_err = final_df['Err.1_x']+final_df['Err.1_y']+final_df['Err.1']
Kills = final_df['Kil_x']+final_df['Kil_y']+final_df['Kil']
Hitting_err = final_df['Err.2_x']+final_df['Err.2_y']+final_df['Err.2']
Att = final_df['Att_x']+final_df['Att_y']+final_df['Att']
Rcv_err = final_df['RcvErr_x']+final_df['RcvErr_y']+final_df['RcvErr']
RcvQualTot = final_df['RcvQualTot_x']+final_df['RcvQualTot_y']+final_df['RcvQualTot']
GamesPlayed = final_df['GamesPlayed_x']+final_df['GamesPlayed_y']+final_df['GamesPlayed']

combined_df = pd.DataFrame({'Players': Players, 'Aces': Aces, 'Srv_err': Srv_err, 'Total_srv': Total_srv, 
                            'RcvCount': RcvCount, 'Rcv_err': Rcv_err, 
                           'RcvQualTot': RcvQualTot, 'Dig': Dig, 'Defense_err': Defense_err, 'Kills': Kills, 
                           'Hitting_err': Hitting_err, 'Att': Att, 'GamesPlayed': GamesPlayed})
combined_df
```
